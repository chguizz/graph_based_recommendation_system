Visualisation

```python
# loading data
y_true = np.loadtxt('saved_model/y_true.txt')
y_hin = np.loadtxt('saved_model/y_pred_HinSAGE.txt')
y_n2v = np.loadtxt('saved_model/y_pred_node2vec.txt')
```

```python
df = pd.DataFrame({'ground truth':y_true, 'HinSAGE':y_hin, 'Node2vec':y_n2v})
df.head()
```

```python
fig, ax = plt.subplots(figsize=(8,6))
sns.histplot(data=df, x="ground truth", bins=30, ax=ax, color='#6aa84f', alpha=.65, edgecolor='#6aa84f', label='ground truth')
sns.histplot(data=df, x="HinSAGE", bins=30, ax=ax, element="step", fill=False, color='navy', alpha=1, label='HinSAGE')
sns.histplot(data=df, x="Node2vec", bins=30, ax=ax, element="step", fill=False, color='firebrick', alpha=1, label='Node2vec')

plt.title('Ranking distributions')
plt.xlabel("ranking")
plt.ylabel("count")

plt.legend()

plt.savefig('figures/ranking_distributions_final.jpg', bbox_inches = 'tight')
plt.show()
```
